## Utilizando o Pycaret para fazer analise previa do Data frame


PyCaret é uma biblioteca de aprendizagem de máquinas de código aberto e de código baixo em Python que automatiza os fluxos de trabalho de aprendizagem de máquinas. É uma ferramenta de aprendizagem de máquina de ponta a ponta e de gerenciamento de modelos que acelera exponencialmente o ciclo de experimentos e o torna mais produtivo.
Em comparação com as outras bibliotecas de aprendizagem de máquinas de código aberto, PyCaret é uma biblioteca alternativa de código baixo que pode ser usada para substituir centenas de linhas de código por apenas algumas linhas. Isto torna os experimentos exponencialmente rápidos e eficientes. PyCaret é essencialmente um invólucro Python que envolve várias bibliotecas e estruturas de aprendizagem de máquinas, tais como Scikit-learn, XGBoost, LightGBM, CatBoost, SpaCy, Optuna, Hyperopt, Ray, e algumas outras.
O design e a simplicidade do PyCaret são inspirados pelo papel emergente dos cientistas de dados. Os Cientistas de Dados emergentes são usuários poderosos que podem realizar tarefas analíticas simples e moderadamente sofisticadas que anteriormente exigiriam mais conhecimentos técnicos.

Neste trabalho, o Pycaret teve função de permitir uma previa se diversos MLs para o dataframe de interesse, este resultado serve como um guia que indica oe melhores desempenhos entre diversos modelos por diferentes métricas.
